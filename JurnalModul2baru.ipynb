## Mendapatkan intensitas tiap piksel

Fungsi getIntensity() digunakan untuk mendapatkan intesitas dari tiap piksel pada citra yang telah diubah ke dalam grayscale. Pada fungsi ini digunakan parameter, yakni citra yang akan dihitung intensitas pikselnya. Apabila diamati setiap nilai pixel akan dimasukkan ke dalam variabel histogram dan apabila ditemui nilai pixel yang sama lagi maka akan ditambahkan satu. diawali dengan pembuatan sebuah array bernama intensity berukuran 256 elemen yang diisi dengan angka nol, dimana tiap indeks mewakili tingkat keabuan dari 0 sampai 255. kemudian lanjut pada iterasi untuk tiap piksel pada gambar yang mengakses masing-masing baris dan kolom. Untuk setiap piksel yang dikunjungi, nilai intensitas piksel tersebut digunakan untuk menambah jumlah hitungan di indeks yang sesuai dalam array intensity. Dengan kata lain, fungsi ini menghitung berapa banyak piksel yang memiliki intensitas tertentu di dalam gambar.

contoh: pada array img[0][0] == 0 maka indeks variabel histogram akan menjadi histogram[0]

```python
def getIntensity(img):
    intensity = np.zeros(256).astype(int)
    #membuat array berisikan 0 dengan besaran 256 dan menampung tipe data integer
    for i in range (img.shape[0]):
        for j in range (img.shape[1]):
            intensity[img[i, j]] += 1
    return intensity
```

## Mendapatkan nilai histogram dari citra asli/awal

fungsi ini digunakan untuk menghitung histogram dari sebuah citra. diawali dengan pembuatan sebuah array bernama intensity berukuran 256 elemen yang diisi dengan angka nol, dimana tiap indeks mewakili tingkat keabuan dari 0 sampai 255. kemudian lanjut pada iterasi untuk tiap piksel pada gambar yang mengakses masing-masing baris dan kolom. setiap sebuah piksel dikunjungi, fungsi akan menambahkan satu elemen ke dalam list pada indeks yang sesuai dengan nilai intensitas piksel tersebut.

```python
# 2. Hitung histogram citra asli
def get_histogram(img):
    hist = np.zeros(256, dtype=int)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            hist[img[i, j]] += 1
    return hist
```

## Mencari nilai distribusi kumulatif

fungsi ini dibuat untuk menghitung nilai distribusi kumulatif, diawali dengan pembuatan array bernama cdf berukuran 256 elemen yang diinisialisasi dengan nilai nol untuk menyimpan nilai cdf. Kemudian, jumlah total piksel dihitung dengan menjumlahkan seluruh elemen pada list. Dengan metode ini, cdf[i] merepresentasikan kemungkinan kumulatif dari intensitas 0 hingga i.

```python
# 3. Hitung distribusi kumulatif (CDF)
def dist_kumulatif(hist):
    cdf = np.zeros(256, dtype=float)
    total_pixels = np.sum(hist)
    cumulative = 0
    for i in range(256):
        cumulative += hist[i]
        cdf[i] = cumulative / total_pixels
    return cdf
```

## Mencari normalisasi nilai distribusi kumulatif

fungsi ini digunakan untuk menormalisasi nilai distribusi kumulatif agar berada dalam rentang 0 sampai 255, 

```python
# 4. Normalisasi CDF ke rentang 0-255
def normalisasi_cdf(cdf):
    cdf_normalized = np.round(cdf * 255).astype(np.uint8)
    return cdf_normalized
```

## Memetakan nilai intensitas piksel

fungsi ini digunakan untuk memetakan nilai intensitas piksel dari sebuah citra berdasarkan hasil ekualisasi dalam citra. Fungsi ini menggunakan 2 parameter, yaitu result yang merupakan array hasil perhitungan baik spesifikasi maupun ekualisasi dan img yang merupakan citra yang digunakan dalam pemetaan hasil perhitungan. Dalam fungsi ini, nilai piksel pada citra img akan digantikan nilainya dengan nilai pada variabel rounded dengan indeks sesuai nilai piksel yang ditunjuk.

```python
# 5. Mapping nilai piksel berdasarkan CDF
def mapping_nilai(img, mapping):
    result = np.zeros_like(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            result[i, j] = mapping[img[i, j]]
    return result
```

## Membuat fungsi final mapping

Fungsi finalMapping() digunakan untuk melakukan pemetaan pada proses spesifikasi. Pada fungsi ini, nilai piksel dari citra asal akan dikurangi dengan tiap nilai intensitas dari citra target untuk mencari nilai terdekat dan ditampung ke variabel temp. Pengurangan ini akan dilakukan sampai indeks terakhir pada citra asal. Setiap perulangan, nilai dari variabe temp akan dibandingkan dengan nilai dari variabel minval. Apabila temp lebih kecil dari minval maka nilai minval akan berganti dengan nilai dari temp begitu juga minj j yang akan diisi dengan nilai j dikurangi 1. Jika perulangan telah selesai, nilai variabel final pada indeks ke-i akan diisi dengan nilai dari minj

```python
def finalMapping(source, target):
    final  = np.zeros(256, dtype=int)
    
    for i in range(256):  # Iterasi untuk setiap nilai intensitas (0-255)
        # Mencari perbedaan absolut minimum antara source[i] dan semua nilai target[j]
        min_val = abs(source[i] - target[0])
        min_idx = 0
        for j in range(256):
            temp = abs(source[i] - target[j])
            if temp < min_val:
                min_val = temp
                min_idx = j  # Menyimpan indeks target dengan perbedaan terkecil
        final[i] = min_idx  # Mapping nilai intensitas
    return final
```

## Membuat fungsi spesifikasi

Fungsi spesifikasi(source, target) bertujuan untuk melakukan spesifikasi histogram dari gambar sumber (source) ke gambar target (target). Pertama-tama, histogram dari kedua gambar dihitung menggunakan fungsi getIntensity. Setelah itu, histogram tersebut diubah menjadi distribusi kumulatif (CDF) melalui fungsi dist_kumulatif. CDF yang dihasilkan kemudian dinormalisasi ke rentang 0–255 dengan normalisasi_cdf. Langkah berikutnya adalah membuat mapping antara nilai-nilai intensitas sumber dan target menggunakan fungsi finalMapping. Terakhir, hasil pemetaan tersebut diterapkan pada gambar sumber melalui fungsi mapping_nilai, dan hasil spesifikasi histogram dikembalikan sebagai output.

```python
# Fungsi untuk spesifikasi histogram citra
def spesifikasi(source, target):
     # Hitung histogram
    hist_source = getIntensity(source)
    hist_target = getIntensity(target)

    # Hitung CDF
    cdf_source = dist_kumulatif(hist_source)
    cdf_target = dist_kumulatif(hist_target)

    # Normalisasi CDF ke 0–255
    norm_cdf_source = normalisasi_cdf(cdf_source)
    norm_cdf_target = normalisasi_cdf(cdf_target)

    # Buat mapping
    mapping = finalMapping(norm_cdf_source, norm_cdf_target)

    # Terapkan mapping ke gambar sumber
    result = mapping_nilai(source, mapping)

    return result
```

## membuat fungsi memotong citra secara horizontal

fungsi ini digunakan untuk membagi sebuah citra menjadi x bagian secara horizontal. diawali dengan mengambil ukuran tinggi dan lebar gambar. kemudian gambar dipotong berdasarkan lebar per bagian dengan perulangan sebanyak x, dimulai dari start_lebar hingga end_lebar. setiap potongan disimpan ke dalam list potongan_gambar

```python
def potong_h(image, x):
    tinggi, lebar = image.shape
    potongan_gambar = []
    for i in range(x):
        start_lebar = (lebar // x) * i
        end_lebar = (lebar // x) * (i + 1)
        bagian = image[:, start_lebar:end_lebar]
        potongan_gambar.append(bagian)
    return potongan_gambar
```
